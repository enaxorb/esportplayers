```python
endpoint = "https://ag1zhv7l0xefmdtk.allegrograph.cloud/repositories/Esport"

query = """
PREFIX wd:      <http://www.wikidata.org/entity/>
PREFIX wdt:     <http://www.wikidata.org/prop/direct/>
PREFIX rdfs:    <http://www.w3.org/2000/01/rdf-schema#>
PREFIX xsd:  <http://www.w3.org/2001/XMLSchema#>


SELECT  ?item 
        (SAMPLE(STR(?itemLabel)) as ?itemLabel) 
        (MAX(?birthDate) as ?birthDate)  
        (MAX(?country) as ?country)
        ( GROUP_CONCAT(DISTINCT ?game; separator=", ") as ?game )
        (GROUP_CONCAT(DISTINCT ?team;  separator=", ") as ?teams)
        (MAX(?careerStart) as ?careerStart)
        (MAX(?twitchFollowers) as ?twitchFollowers)
        (MAX(?competitionWon) as ?competitionWon)
WHERE {
    GRAPH <https://github.com/enaxorb/esportplayers/blob/main/graphs/wikidata-imported-data.md> {        
    ?item wdt:P569 ?birthDate;
        wdt:P27 ?country;
        wdt:P2416 ?game;
    
    OPTIONAL { ?item wdt:P54    ?team }
    OPTIONAL { ?item wdt:P2031  ?careerStart }
    OPTIONAL { ?item wdt:P8687  ?twitchFollowers }
    OPTIONAL { ?item wdt:P2522  ?competitionWon }

    OPTIONAL {?item rdfs:label ?itemLabel .}
  }  
}
GROUP BY ?item
"""
```

```python
#préparation des données
qr = spqf.get_json_sparql_result(endpoint, query)
result = spqf.sparql_result_to_list(qr)

dfg = pd.DataFrame(result)
dfg.columns = ['personUri','itemLabel','birthYear','country','game', 'team','careerStart','twitchFollowers','competitionWon']
#Enlever les NA
dfg = dfg.applymap(lambda x: x.strip() if isinstance(x, str) else x)
dfg.replace(r'^\s*$', pd.NA, regex=True, inplace=True)

#Nettoyage
def clean_id(s):
    if isinstance(s,str) and s.startswith('http'):
        return s.rsplit('/', 1) [-1]
    return s
dfg['game_clean'] = dfg['game'].apply(clean_id)

def split_games(s):
    if pd.isna(s) or s =="":
        return[]
    return [g.strip() for g in str(s).split(",") if g.strip()]
dfg['games_list'] = dfg['game_clean'].apply(split_games)

dfg['games_list']=dfg['game_clean'].apply(split_games)
edges_player_game = (
    dfg[['itemLabel', 'games_list']]
      .explode('games_list')
      .dropna(subset=['games_list'])
)

print(edges_player_game.head())
print(f"Nombre total de liens joueur-jeu : {len(edges_player_game)}")
```

```python
#Graphe bipartite joueurs/jeux
B = nx.Graph()
B.add_nodes_from(edges_player_game['itemLabel'], bipartite=0)
B.add_nodes_from(edges_player_game['games_list'], bipartite=1)
B.add_edges_from(edges_player_game[['itemLabel', 'games_list']].to_records(index=False))

print(f"Nœuds : {B.number_of_nodes()}, Arêtes : {B.number_of_edges()}")

#Projection jeu/jeu
games_nodes = [n for n, d in B.nodes(data=True) if d['bipartite'] == 1]
G_game = nx.bipartite.weighted_projected_graph(B, games_nodes)

print(f"Jeux : {G_game.number_of_nodes()}, Arêtes : {G_game.number_of_edges()}")

#Jeux "pivots"
deg = dict(G_game.degree(weight='weight'))
betw = nx.betweenness_centrality(G_game, weight='weight', normalized=True)

top_deg = sorted(deg.items(), key=lambda x: x[1], reverse=True)[:10]
top_betw = sorted(betw.items(), key=lambda x: x[1], reverse=True)[:10]
```

```python
#Communauté
from networkx.algorithms.community import greedy_modularity_communities
C = list(greedy_modularity_communities(G_game, weight='weight'))
comm_id = {}
for i, com in enumerate(C):
    for n in com:
        comm_id[n] = i
nx.set_node_attributes(G_game, comm_id, 'community')
```

```python
#Visualisation
thr = 1
H = G_game.edge_subgraph([(u,v) for u,v,d in G_game.edges(data=True) if d['weight']>=thr]).copy()
pos = nx.spring_layout(H, seed=42, k=0.3)

plt.figure(figsize=(11,9))
node_colors = [H.nodes[n].get('community', -1) for n in H.nodes()]
node_sizes  = [200 + 50*deg.get(n,0) for n in H.nodes()]
edge_w      = [0.4 + 0.2*d['weight'] for _,_,d in H.edges(data=True)]

nx.draw_networkx_nodes(H, pos, node_size=node_sizes, node_color=node_colors, cmap=plt.cm.tab20, alpha=0.85)
nx.draw_networkx_edges(H, pos, width=edge_w, alpha=0.35)
nx.draw_networkx_labels(H, pos, font_size=8)
plt.title("Structure des communautés de jeux vidéo basées sur les joueurs en commun")
plt.axis('off'); plt.show()
```
